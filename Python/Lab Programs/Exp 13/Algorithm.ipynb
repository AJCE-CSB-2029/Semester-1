### **Algorithm: Reverse a Number**

1. **Start**
2. **Input** a number **n**
3. **Store** the value of **n** in **original**
4. **Initialize** `reverse = 0`
5. **While** `n > 0`, repeat:

   * Extract the last digit: `digit = n % 10`
   * Append digit to reverse: `reverse = reverse * 10 + digit`
   * Remove the last digit: `n = n // 10`
6. **Display** the reversed number
7. **Stop**
