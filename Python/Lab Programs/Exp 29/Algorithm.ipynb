### **Algorithm: Dictionary Creation, Access, and Update**

1. **Start**
2. **Initialize** an empty dictionary → `student = {}`
3. **Input** how many items to add → `n`
4. **For** each of the `n` items:

   * Input a `key`
   * Input its corresponding `value`
   * **Store** the pair in the dictionary → `student[key] = value`
5. **Display** the dictionary

---

#### **Accessing Dictionary Items**

6. Input a key → `key_input`
7. **If** `key_input` exists in `student`:

   * Display the value using `student[key_input]`
8. Else, display **“Invalid Key!”**

---

#### **Accessing Using `get()` Method**

9. Input a key → `get_key`
10. **If** `get_key` exists in dictionary:

    * Display value using `student.get(get_key)`
11. Else, display **“Invalid Key!”**

---

#### **Updating Values**

12. Input the key to modify → `change_key`
13. Input the new value → `change_value`
14. **If** `change_key` exists:

    * Update value → `student[change_key] = change_value`
15. Else, display **“Invalid Key!”**

---

16. **Display** the updated dictionary
17. **Stop**
