### **Algorithm: Create, concatenate and access sub-string**

1. **Start**

2. **Input** the first string → `str1`
3. **Input** the second string → `str2`
4. **Concatenate** the two strings → `concat_str = str1 + str2`
5. **Display** the concatenated string
6. **Input** the starting index → `start`
7. **Input** the ending index → `end`
8. **Extract** substring → `concat_str[start : end]`
9. **Display** the substring
10. **Stop**
