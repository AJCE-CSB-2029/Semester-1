### **Algorithm: Create a dictionary with opposite mapping**

1. **Start**
2. **Initialize** the original dictionary → `x`
3. **Create** an empty dictionary → `opposite`
4. **For each key** in dictionary `x`:

   * Get the corresponding value → `value = x[key]`
   * **Store** the reversed mapping in `opposite` → `opposite[value] = key`
5. **Display** the original dictionary
6. **Display** the reversed dictionary (`opposite`)
7. **Stop**
