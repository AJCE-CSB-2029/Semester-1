### **Algorithm: Armstrong Number Checker Using Function**

1. **Start**
2. **Define function `armstrong(num)`**

   * Convert `num` to string and find number of digits → `length`
   * Set `temp = num`
   * Initialize `total = 0`
   * While `temp > 0`

     * Extract digit: `digit = temp % 10`
     * Add `digit ** length` to `total`
     * Reduce number: `temp = temp // 10`
   * Return `total`
3. **Input** a number → `n`
4. **Call** `armstrong(n)` and store result in `arm`
5. **If** `arm == n`

   * Display **“It is an Armstrong number”**
6. **Else**

   * Display **“It is not an Armstrong number”**
7. **Stop**
