### **Algorithm: Reverse a String (Built-in + User-Defined Function)**

### **Part 1: Using Built-in Function (`reversed()`)**

1. **Start**

2. **Input** a string → `s`
3. **Apply built-in function** `reversed(s)` to get characters in reverse order
4. **Join** the reversed characters using `"".join()` → store in `rev`
5. **Display** the reversed string

---

### **Part 2: Using User-Defined Function**

6. **Define** a function `reverse_string(s)`

   * Initialize `rev = ""`
   * For each character `ch` in `s`:

     * Prepend the character: `rev = ch + rev`
   * Return `rev`
7. **Call** the function with the input string → `reverse_string(s)`
8. **Display** the reversed string returned by the function
9. **Stop**
