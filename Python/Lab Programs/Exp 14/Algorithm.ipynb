### **Algorithm: Simple Calculator using Function**

1. **Start**
2. **Display** the menu options:

   * 1. Addition
   * 2. Subtraction
   * 3. Multiplication
   * 4. Division
3. **Input** the user's choice → store in **choice**
4. **Input** two numbers → store in **num1** and **num2**
5. **If** `choice == 1`, then

   * Compute `result = num1 + num2`
   * Display **result**
6. **Else if** `choice == 2`, then

   * Compute `result = num1 - num2`
   * Display **result**
7. **Else if** `choice == 3`, then

   * Compute `result = num1 * num2`
   * Display **result**
8. **Else if** `choice == 4`, then

   * If `num2 == 0`, display **"Division by zero not possible"**
   * Else compute `result = num1 / num2`
   * Display **result**
9. **Else**, display **"Invalid choice"**
10. **Stop**

### **Algorithm: Simple Calculator using Function (with loop)**

1. **Start**
2. **Repeat Forever (loop):**

   1. Display menu:

      * 1 Addition
      * 2 Subtraction
      * 3 Multiplication
      * 4 Division
      * 5 Exit
   2. Input **choice**
   3. If **choice = 5**, display “Exiting…” and **stop**
   4. Input two numbers: **num1**, **num2**
   5. If **choice = 1**, compute and display **num1 + num2**
   6. Else if **choice = 2**, compute and display **num1 - num2**
   7. Else if **choice = 3**, compute and display **num1 × num2**
   8. Else if **choice = 4**,

      * If **num2 = 0**, display “Division not possible”
      * Else compute and display **num1 ÷ num2**
   9. Else display “Invalid choice”
3. **Stop**
