### **Algorithm: Create a List of Numbers Less Than a Given Value**

1. **Start**
2. **Input** a list of integers → `master_list`
3. **Input** a number → `limit`
4. **Initialize** an empty list → `new_list`
5. **For each** number `num` in `master_list`:

   * If `num < limit`, append `num` to `new_list`
6. **Display** `new_list`
7. **Stop**
