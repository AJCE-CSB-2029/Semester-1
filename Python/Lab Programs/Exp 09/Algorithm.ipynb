### **Algorithm: Factorial of a Number**

1. **Start**

2. **Input** a number **n**
3. **If** `n < 0`, then

   * Display **“Factorial is not possible for a negative number.”**
   * **Stop**
4. **Else if** `n == 0`, then

   * Display **“Factorial = 1”**
   * **Stop**
5. **Else**

   * Initialize `fact = 1`
   * Initialize `i = 1`
6. **While** `i ≤ n`, repeat:

   * `fact = fact * i`
   * Increment `i`
7. **Display** `fact` as the factorial of **n**
8. **Stop**
