### **Algorithm: Count the frequency of words in a string using dictionary**

1. **Start**
2. **Input** a string → `string`
3. **Split** the string into words → `words = string.split()`
4. **Initialize** an empty dictionary → `freq = {}`
5. **For each word** `w` in `words`:

   * **If** `w` exists in dictionary `freq`:

     * Increment its count → `freq[w] = freq[w] + 1`
   * **Else**:

     * Add the word to dictionary with count 1
6. **Display** the dictionary `freq` containing word frequencies
7. **Stop**
