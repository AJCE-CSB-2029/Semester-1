### **Algorithm: Sum of Two Numbers Using a Function**

1. **Start**
2. **Define a function** `sum_of_two(a, b)`

   * Compute `sum = a + b`
   * Return `sum`
3. **Input** the first number → `a`
4. **Input** the second number → `b`
5. **Call** the function with arguments `a` and `b`
6. **Store** returned result in `sum`
7. **Display** the value of `sum`
8. **Stop**
