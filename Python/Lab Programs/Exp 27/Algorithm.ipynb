### **Algorithm: Find the Third Largest Element in a List**

1. **Start**
2. **Input** a list of numbers → `list_0`
3. **Initialize** an empty list → `unique_list`
4. **For each** element `num` in `list_0`:

   * If `num` is **not** in `unique_list`, append it
5. **Sort** `unique_list` in **descending order**
6. **If** the length of `unique_list` is **less than 3**:

   * Display **“Not enough elements”**
7. **Else**:

   * Set `third_largest = unique_list[2]`
   * Display `third_largest`
8. **Stop**
