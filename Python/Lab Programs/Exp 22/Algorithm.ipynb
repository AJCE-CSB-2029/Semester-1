### **Algorithm 1: Word Count by Reading the Whole File at Once**

1. **Start**
2. **Open** the file **python.txt** in **read mode**
3. **Read** the entire content of the file into a variable `content`
4. **Close** the file
5. **Initialize** a counter `count = 0`
6. **Split** `content` into words using `content.split()`
7. **For each word** in the splitted list:

   * Increment the counter `count = count + 1`
8. **Display** the value of `count`
9. **Stop**
