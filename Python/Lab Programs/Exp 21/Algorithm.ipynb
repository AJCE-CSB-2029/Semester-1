### **Algorithm: File Write and Read Operation**

1. **Start**

2. **Open** a file named **python.txt** in **write mode ('w')**
3. **Write** the text:
   `"Hello world \nThis is a python program"`
   into the file
4. **Close** the file
5. **Open** the file **python.txt** again in **read mode ('r')**
6. **Read** the entire content of the file and store it in a variable
7. **Close** the file
8. **Display** the content on the screen
9. **Stop**
