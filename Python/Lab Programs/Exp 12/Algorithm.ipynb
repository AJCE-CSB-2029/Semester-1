### **Algorithm: Armstrong Number Checker**

1. **Start**
2. **Input** a number **num**
3. **Convert** `num` to string and find its length â†’ store in **length**
4. **Store** `num` in a temporary variable **temp**
5. **Initialize** `sum = 0`
6. **While** `temp > 0`, repeat:

   * Extract last digit: `digit = temp % 10`
   * Compute digit raised to power of length: `digit ** length`
   * Add to sum: `sum = sum + (digit ** length)`
   * Remove last digit: `temp = temp // 10`
7. **If** `sum == num`, then

   * Display **"It is Armstrong"**
8. **Else**

   * Display **"It is not Armstrong"**
9. **Stop**
