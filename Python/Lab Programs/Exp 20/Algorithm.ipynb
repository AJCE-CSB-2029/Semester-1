### **Algorithm: Palindrome and Symmetrical String Checker**

#### **1. Palindrome Check**

1. **Start**
2. **Input** a string → `s`
3. **Reverse** the string using slicing → `rev = s[::-1]`
4. **If** `s == rev`, display **"Palindrome"**
5. Else display **"Not Palindrome"**

---

#### **2. Symmetry Check**

6. Find the **half length** → `half = len(s) // 2`
7. **If** length of string is even:

   * `first_half = s[:half]`
   * `second_half = s[half:]`
8. **Else** (length is odd):

   * `first_half = s[:half]`
   * `second_half = s[half + 1:]`  (skipping the middle character)
9. **If** `first_half == second_half`, display **"Symmetrical"**
10. Else display **"Not Symmetrical"**
11. **Stop**
