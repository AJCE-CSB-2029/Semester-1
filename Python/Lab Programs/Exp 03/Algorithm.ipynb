### **Algorithm: Swap Two Numbers Using a Temporary Variable**
1. **Start**

2. **Input** a and b
4. **Store** the value of **a** in a temporary variable → `temp = a`
5. **Assign** the value of **b** to **a** → `a = b`
6. **Assign** the value stored in **temp** to **b** → `b = temp`
7. **Display** the swapped values of **a** and **b**
8. **Stop**
