### **Algorithm: Copy Contents From One File to Another**

1. **Start**

2. **Open** the file **python.txt** in **read mode**
3. **Read** the content of the file into a variable `content`
4. **Close** the file
5. **Open** a new file **python_1.txt** in **write mode**
6. **Write** the content stored in `content` into this new file
7. **Close** the new file
8. **Open** **python_1.txt** again in **read mode**
9. **Read** the copied content into a variable `copied_content`
10. **Close** the file
11. **Display** a message saying the file is copied, along with the copied content
12. **Stop**
