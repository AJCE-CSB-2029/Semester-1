### **Algorithm: Create, Append, and Insert Elements in a List**

1. **Start**
2. **Input** the number of elements → `n`
3. **Create** a list of size `n` initialized with zeros → `my_list = [0] * n`
4. **For** each index `i` from 0 to `n−1`:

   * Input an element
   * Store it in `my_list[i]`
5. **Display** the list
6. **Input** an element → `append_num`
7. **Append** this element to the list using `append()`
8. **Display** the updated list
9. **Input** a number → `insert_num`
10. **Input** the index → `index`
11. **Insert** the number at the given index using `insert(index, insert_num)`
12. **Display** the updated list
13. **Stop**
