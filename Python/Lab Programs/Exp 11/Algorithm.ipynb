### **Algorithm: Palindrome Checker for a Number (Using while loop)**

1. **Start**
2. **Input** a number **n**
3. **Store** the value of **n** in **original**
4. **Initialize** `reverse = 0`
5. **While** `n > 0`, repeat:

   * Extract last digit: `digit = n % 10`
   * Append digit to reverse: `reverse = reverse * 10 + digit`
   * Remove last digit from n: `n = n // 10`
6. **If** `original == reverse`, then

   * Display **"It is a Palindrome"**
7. **Else**

   * Display **"It's not a Palindrome"**
8. **Stop**
