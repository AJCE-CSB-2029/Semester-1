### **Algorithm: Fibonacci Sequence (Using Temporary Variable)**

1. **Start**

2. **Input** a number **n**
3. **Initialize**

   * `a = 0`
   * `b = 1`
   * `i = 0`
4. **While** `i < n`, repeat:

   * Display **a**
   * Store the value of **a** in `temp`
   * Assign `a = b`
   * Assign `b = temp + b`
   * Increment `i`
5. **Stop**
