### **Algorithm: Prime Number Checker Using a Function + Printing Primes in a Range**

#### **Function: is_prime(number)**

1. **Define** a function `is_prime(number)`
2. **If** `number ≤ 1`

   * Return **False**
3. **For** each `factor` from 2 to √`number`, do:

   * **If** `number % factor == 0`, then

     * Return **False**
4. **If no factor divides the number**, return **True**

---

#### **Main Program**

1. **Start**
2. **Input** the starting number → `start`
3. **Input** the ending number → `end`
4. **For** each number `value` from `start` to `end`:

   * **Call** the function `is_prime(value)`
   * **If** it returns **True**, display `value`
5. **Stop**
