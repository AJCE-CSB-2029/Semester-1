### **Algorithm: Find the Largest of Three Numbers**

1. **Start**
2. **Input three numbers** **a**, **b**, and **c**
3. **If** `a > b` **and** `a > c`, then

   * **a** is the largest
4. **Else if** `b > c`, then

   * **b** is the largest
5. **Else**

   * **c** is the largest
6. **Display** the largest number
7. **Stop**
